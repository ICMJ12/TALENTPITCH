```python
import pandas as pd
from sqlalchemy import create_engine
import sqlalchemy
from datetime import datetime
```

```python
# Leer los archivos .csv
df_challenges = pd.read_csv("challenges - challenges.csv")
df_profiles = pd.read_csv("profiles - Hoja 1.csv")
df_resumes = pd.read_csv("resumes - resumes.csv")
df_users = pd.read_csv("users - users.csv")
```

```python
df_challenges
```

```python
# Cambiar el tipo de dato, fecha
df_challenges['created_at'] = pd.to_datetime(df_challenges['created_at'])
df_profiles['updated_at'] = pd.to_datetime(df_profiles['updated_at'])
df_profiles['created_at'] = pd.to_datetime(df_profiles['created_at'])
df_resumes['created_at'] = pd.to_datetime(df_resumes['created_at'])
df_users['created_at'] = pd.to_datetime(df_users['created_at'])
df_users['updated_at'] = pd.to_datetime(df_users['updated_at'])
```

```python
# Cambiar la estructura de la fecha
df_challenges['created_at'] = df_challenges['created_at'].dt.strftime('%Y-%m-%d')
df_profiles['updated_at'] = df_profiles['updated_at'].dt.strftime('%Y-%m-%d')
df_profiles['created_at'] = df_profiles['created_at'].dt.strftime('%Y-%m-%d')
df_resumes['created_at'] = df_resumes['created_at'].dt.strftime('%Y-%m-%d')
df_users['created_at'] = df_users['created_at'].dt.strftime('%Y-%m-%d')
df_users['updated_at'] = df_users['updated_at'].dt.strftime('%Y-%m-%d')
```

```python
# Conexión a MySQL
user = 'root'       # Reemplaza con tu usuario
password = 'Camilo12*'   # Reemplaza con tu contraseña
host = 'localhost'           # Por ejemplo, 'localhost' o la IP de tu servidor MySQL
database = 'talentpitch'   # El nombre de tu base de datos

# Crear la conexión
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')

# Cargar los DataFrames en MySQL
df_challenges.to_sql('challenges', con=engine, if_exists='replace', index=False)
df_profiles.to_sql('profiles', con=engine, if_exists='replace', index=False)
df_resumes.to_sql('resumes', con=engine, if_exists='replace', index=False)
df_users.to_sql('users', con=engine, if_exists='replace', index=False)

print("Datos cargados en MySQL con éxito.")
```
